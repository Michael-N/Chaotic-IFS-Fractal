```python
#imports
from IteratedFunctionSystem import IFS
from ChaoticIFSBuilder import ChaoticIFSBuilder
from constants import constants
import numpy as np, pandas as pd, datashader as ds
from datashader import transfer_functions as tf
from datashader.colors import inferno, viridis
from numba import jit
import numpy.random
from colorcet import palette
import math
from decimal import *
import bokeh.plotting as bp
from datashader.bokeh_ext import InteractiveImage
import tqdm
from PlottingUtil import PlottingUtil
```

```python
#a gui for asking the user for function settings: asks the user for settings 
newSystem = ChaoticIFSBuilder(constants,"goldendragon")
settings = newSystem.run()
function_set = settings[0]
points_quantity = settings[1]
speedup_mode = settings[2]

#Setup the IFS (SPEEDUP does not take into account theta and instead asks theta to be calculated into the stretch matrix beforehand)
if(speedup_mode):
    func = IFS.chaoticQuickAffineGenerator(function_set)
else:
    func = IFS.chaoticAffineGenerator(function_set)
    
#Preform the IFS calculations...
point_data = IFS.run(func,max_points=points_quantity)

#add to the points by clustering points around eachother by some differntial
additional_point_data = [[],[]]
delta=0.01
for i in range(0,len(point_data)):
    x1 = point_data[0][i] +delta
    x2 = x1 -2*delta
    y1 = point_data[1][i] +delta
    y2 = y1 - 2*delta
    additional_point_data[0].append(x1)
    additional_point_data[0].append(x2)
    additional_point_data[0].append(x1)
    additional_point_data[0].append(x2)
    additional_point_data[1].append(y1)
    additional_point_data[1].append(y2)
    additional_point_data[1].append(y2)
    additional_point_data[1].append(y1)
n_point_data=[[],[]]
n_point_data[0] = list(point_data[0]) + list(additional_point_data[0])
n_point_data[1] = list(point_data[1]) + list(additional_point_data[1])

#Use the plotting Util to plot the data
pu = PlottingUtil(n_point_data)
pu.interactive_plot()
```
